```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.special import beta as beta_func
from scipy.special import gamma as gamma_func
from scipy.optimize import minimize
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import warnings
warnings.filterwarnings('ignore')

# ЗАВАНТАЖЕННЯ ДАНИХ
df = pd.read_csv(r'C:\Users\user\Downloads\KMA TestTask Product Analyst - Task2.csv')
print("="*80)
print("LTV PREDICTION ANALYSIS - PROBABILISTIC MODELS (sBG + Gamma/BdW) — Hanna Todorova")
print("="*80)

# 1. ПІДГОТОВКА ДАНИХ
print("\n1. ПІДГОТОВКА ДАНИХ")
print("-"*80)

df['event_timestamp'] = pd.to_datetime(df['event_timestamp'])
df['first_purchase_time'] = pd.to_datetime(df['first_purchase_time'])
df['cohort_week'] = pd.to_datetime(df['cohort_week'])
df['price'] = df['product_id'].str.extract(r'(\d+\.?\d*)').astype(float)

print(f"Загальна кількість транзакцій: {len(df):,}")
print(f"Кількість унікальних користувачів: {df['user_id'].nunique():,}")
print(f"Період даних: {df['cohort_week'].min()} - {df['cohort_week'].max()}")
print(f"Максимальний lifetime_weeks: {df['lifetime_weeks'].max()}")

# 2. РОЗРАХУНОК ФАКТИЧНИХ МЕТРИК
print("\n\n2. РОЗРАХУНОК ФАКТИЧНИХ МЕТРИК ПО КОГОРТАХ")
print("-"*80)

# Для кожної когорти розраховуємо метрики
cohort_metrics = []

for cohort in df['cohort_week'].unique():
    cohort_users = df[df['cohort_week'] == cohort]
    cohort_size = cohort_users['user_id'].nunique()
    
    # По кожному тижню
    for week in range(int(cohort_users['lifetime_weeks'].max()) + 1):
        # Користувачі активні на цьому тижні
        active_users_week = cohort_users[cohort_users['lifetime_weeks'] == week]['user_id'].nunique()
        
        # Кумулятивний revenue до цього тижня включно
        cum_revenue = cohort_users[cohort_users['lifetime_weeks'] <= week]['price'].sum()
        
        # Revenue саме на цьому тижні
        week_revenue = cohort_users[cohort_users['lifetime_weeks'] == week]['price'].sum()
        
        cohort_metrics.append({
            'cohort_week': cohort,
            'lifetime_weeks': week,
            'cohort_size': cohort_size,
            'active_users': active_users_week,
            'retention': active_users_week / cohort_size if cohort_size > 0 else 0,
            'cumulative_revenue': cum_revenue,
            'week_revenue': week_revenue,
            'ltv': cum_revenue / cohort_size if cohort_size > 0 else 0,
            'arpu_week': week_revenue / cohort_size if cohort_size > 0 else 0
        })

cohort_df = pd.DataFrame(cohort_metrics)

print("\nПриклад метрик (перша когорта, перші 8 тижнів):")
sample_cohort = cohort_df['cohort_week'].min()
sample_data = cohort_df[cohort_df['cohort_week'] == sample_cohort].head(8)
print(sample_data[['lifetime_weeks', 'retention', 'ltv', 'arpu_week']].to_string(index=False))

# 3. ЙМОВІРНІСНІ МОДЕЛІ
print("\n\n3. ЙМОВІРНІСНІ МОДЕЛІ ПРОГНОЗУВАННЯ")
print("-"*80)
print("Використовуємо дві моделі:")
print("1. sBG (Shifted Beta-Geometric) - для моделювання retention/churn")
print("2. Gamma/BdW - для моделювання монетизації")
print("\nПеревага: враховують ймовірність churn кожного тижня")

# sBG модель для retention
def sbg_retention(t, alpha, beta):
    """
    sBG (Shifted Beta-Geometric) модель для retention
    P(активний на тижні t) = beta(alpha, beta + t) / beta(alpha, beta)
    """
    if t == 0:
        return 1.0
    return beta_func(alpha, beta + t) / beta_func(alpha, beta)

def sbg_survival(t, alpha, beta):
    """Ймовірність дожити до тижня t"""
    return np.array([sbg_retention(ti, alpha, beta) for ti in t])

# Gamma/BdW для монетизації
def gamma_revenue_per_period(t, r, alpha_rev, scale):
    """
    Очікуваний revenue на period для активних користувачів
    Використовує Gamma розподіл
    """
    return scale * (alpha_rev / (alpha_rev + t))

# Функція для фітування sBG моделі
def fit_sbg_model(retention_data):
    """
    Фітуємо sBG модель до даних retention
    retention_data: dict з ключами - номер тижня, значення - retention rate
    """
    weeks = np.array(list(retention_data.keys()))
    actual_retention = np.array(list(retention_data.values()))
    
    def loss(params):
        alpha, beta = params
        if alpha <= 0 or beta <= 0:
            return 1e10
        predicted = sbg_survival(weeks, alpha, beta)
        return np.sum((actual_retention - predicted) ** 2)
    
    # Початкові параметри
    result = minimize(loss, [1.0, 1.0], bounds=[(0.01, 10), (0.01, 10)], method='L-BFGS-B')
    
    if result.success:
        return result.x
    else:
        return [1.0, 1.0]

# Функція для фітування revenue моделі
def fit_revenue_model(arpu_data, retention_data):
    """
    Фітуємо Gamma модель до даних ARPU
    """
    weeks = np.array(list(arpu_data.keys()))
    actual_arpu = np.array(list(arpu_data.values()))
    actual_retention = np.array([retention_data.get(w, 0) for w in weeks])
    
    # Нормалізуємо ARPU на retention (revenue per active user)
    revenue_per_active = np.zeros_like(actual_arpu)
    for i, (arpu, ret) in enumerate(zip(actual_arpu, actual_retention)):
        if ret > 0:
            revenue_per_active[i] = arpu / ret
        else:
            revenue_per_active[i] = 0
    
    def loss(params):
        r, alpha_rev, scale = params
        if r <= 0 or alpha_rev <= 0 or scale <= 0:
            return 1e10
        predicted = np.array([gamma_revenue_per_period(w, r, alpha_rev, scale) for w in weeks])
        mask = revenue_per_active > 0
        if not np.any(mask):
            return 1e10
        return np.sum((revenue_per_active[mask] - predicted[mask]) ** 2)
    
    # Початкові параметри
    avg_revenue = np.mean(revenue_per_active[revenue_per_active > 0]) if np.any(revenue_per_active > 0) else 1.0
    result = minimize(loss, [1.0, 1.0, avg_revenue], 
                     bounds=[(0.01, 5), (0.01, 5), (0.01, avg_revenue * 3)], 
                     method='L-BFGS-B')
    
    if result.success:
        return result.x
    else:
        return [1.0, 1.0, avg_revenue]

# 4. ТРЕНУВАННЯ МОДЕЛЕЙ ПО КОГОРТАХ
print("\n\n4. ПОБУДОВА МОДЕЛЕЙ ПО КОГОРТАХ")
print("-"*80)

predictions = []
model_params = {}
min_weeks = 4

for cohort in cohort_df['cohort_week'].unique():
    cohort_data = cohort_df[cohort_df['cohort_week'] == cohort].copy()
    cohort_data = cohort_data.sort_values('lifetime_weeks')
    max_week = cohort_data['lifetime_weeks'].max()
    
    if max_week >= min_weeks:
        # Готуємо дані для фітування
        retention_dict = dict(zip(cohort_data['lifetime_weeks'], cohort_data['retention']))
        arpu_dict = dict(zip(cohort_data['lifetime_weeks'], cohort_data['arpu_week']))
        
        try:
            # Фітуємо sBG для retention
            alpha, beta = fit_sbg_model(retention_dict)
            
            # Фітуємо Gamma для revenue
            r, alpha_rev, scale = fit_revenue_model(arpu_dict, retention_dict)
            
            model_params[cohort] = {
                'alpha': alpha,
                'beta': beta,
                'r': r,
                'alpha_rev': alpha_rev,
                'scale': scale
            }
            
            cohort_size = cohort_data['cohort_size'].iloc[0]
            
            # Прогноз на 52 тижні ПІСЛЯ останньої когорти
            weeks_to_predict = np.arange(0, 53)
            
            for week in weeks_to_predict:
                # Прогноз retention
                retention_pred = sbg_retention(week, alpha, beta)
                
                # Прогноз revenue per active user
                revenue_per_active = gamma_revenue_per_period(week, r, alpha_rev, scale)
                
                # ARPU = retention * revenue_per_active
                arpu_pred = retention_pred * revenue_per_active
                
                # Кумулятивний LTV (сума всіх попередніх тижнів)
                if week == 0:
                    ltv_pred = arpu_pred
                else:
                    prev_ltv = [p['ltv_predicted'] for p in predictions 
                               if p['cohort_week'] == cohort and p['lifetime_weeks'] == week - 1]
                    ltv_pred = (prev_ltv[0] if prev_ltv else 0) + arpu_pred
                
                pred_record = {
                    'cohort_week': cohort,
                    'lifetime_weeks': week,
                    'cohort_size': cohort_size,
                    'retention_predicted': retention_pred,
                    'arpu_predicted': arpu_pred,
                    'ltv_predicted': ltv_pred
                }
                
                # Додаємо фактичні значення
                actual_row = cohort_data[cohort_data['lifetime_weeks'] == week]
                if not actual_row.empty:
                    pred_record['retention_actual'] = actual_row['retention'].values[0]
                    pred_record['arpu_actual'] = actual_row['arpu_week'].values[0]
                    pred_record['ltv_actual'] = actual_row['ltv'].values[0]
                else:
                    pred_record['retention_actual'] = np.nan
                    pred_record['arpu_actual'] = np.nan
                    pred_record['ltv_actual'] = np.nan
                
                predictions.append(pred_record)
                
        except Exception as e:
            print(f"Помилка для когорти {cohort.strftime('%Y-%m-%d')}: {e}")

predictions_df = pd.DataFrame(predictions)
print(f"\nКількість когорт з моделями: {len(model_params)}")
print(f"Прогноз побудовано до {predictions_df['lifetime_weeks'].max()} тижнів")

# Виводимо параметри моделей
print("\nПриклад параметрів моделі (перша когорта):")
first_cohort = list(model_params.keys())[0]
params = model_params[first_cohort]
print(f"  sBG: α={params['alpha']:.3f}, β={params['beta']:.3f}")
print(f"  Gamma/BdW: r={params['r']:.3f}, α_rev={params['alpha_rev']:.3f}, scale=${params['scale']:.2f}")

# 5. ОЦІНКА ЯКОСТІ
print("\n\n5. ОЦІНКА ЯКОСТІ МОДЕЛІ")
print("-"*80)

validation_data = predictions_df[predictions_df['ltv_actual'].notna()].copy()

if len(validation_data) > 0:
    mae = mean_absolute_error(validation_data['ltv_actual'], validation_data['ltv_predicted'])
    rmse = np.sqrt(mean_squared_error(validation_data['ltv_actual'], validation_data['ltv_predicted']))
    r2 = r2_score(validation_data['ltv_actual'], validation_data['ltv_predicted'])
    
    non_zero = validation_data[validation_data['ltv_actual'] > 0]
    if len(non_zero) > 0:
        mape = np.mean(np.abs((non_zero['ltv_actual'] - non_zero['ltv_predicted']) / 
                              non_zero['ltv_actual'])) * 100
    else:
        mape = 0
    
    print(f"Mean Absolute Error (MAE): ${mae:.2f}")
    print(f"Root Mean Square Error (RMSE): ${rmse:.2f}")
    print(f"R² Score: {r2:.4f}")
    print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
    
    # Окремо для retention
    retention_valid = validation_data[validation_data['retention_actual'].notna()]
    if len(retention_valid) > 0:
        retention_mae = mean_absolute_error(retention_valid['retention_actual'], 
                                            retention_valid['retention_predicted'])
        print(f"\nRetention MAE: {retention_mae*100:.2f}%")
else:
    mae = rmse = r2 = mape = 0

# 6. ВІЗУАЛІЗАЦІЯ
print("\n\n6. ВІЗУАЛІЗАЦІЯ")
print("-"*80)

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('LTV Prediction Analysis - Probabilistic Models (sBG + Gamma/BdW)', 
             fontsize=16, fontweight='bold')

# 6.1. Retention кривих
ax1 = axes[0, 0]
cohorts_to_plot = predictions_df['cohort_week'].unique()[:5]

for i, cohort in enumerate(cohorts_to_plot):
    cohort_pred = predictions_df[predictions_df['cohort_week'] == cohort]
    actual_data = cohort_pred[cohort_pred['retention_actual'].notna()]
    
    color = plt.cm.Set2(i)
    ax1.scatter(actual_data['lifetime_weeks'], actual_data['retention_actual'] * 100, 
               alpha=0.7, s=50, color=color, label=f'{cohort.strftime("%m-%d")}')
    ax1.plot(cohort_pred['lifetime_weeks'], cohort_pred['retention_predicted'] * 100, 
            '--', alpha=0.8, linewidth=2, color=color)

ax1.set_xlabel('Lifetime (Weeks)', fontweight='bold')
ax1.set_ylabel('Retention (%)', fontweight='bold')
ax1.set_title('Retention: ● Actual  - - Predicted (sBG)', fontweight='bold')
ax1.legend(fontsize=8, loc='upper right')
ax1.grid(True, alpha=0.3)
ax1.set_ylim(0, 105)

# 6.2. LTV кривих
ax2 = axes[0, 1]
for i, cohort in enumerate(cohorts_to_plot):
    cohort_pred = predictions_df[predictions_df['cohort_week'] == cohort]
    actual_data = cohort_pred[cohort_pred['ltv_actual'].notna()]
    
    color = plt.cm.Set2(i)
    ax2.scatter(actual_data['lifetime_weeks'], actual_data['ltv_actual'], 
               alpha=0.7, s=50, color=color, label=f'{cohort.strftime("%m-%d")}')
    ax2.plot(cohort_pred['lifetime_weeks'], cohort_pred['ltv_predicted'], 
            '--', alpha=0.8, linewidth=2, color=color)

ax2.set_xlabel('Lifetime (Weeks)', fontweight='bold')
ax2.set_ylabel('LTV ($)', fontweight='bold')
ax2.set_title('LTV: ● Actual  - - Predicted', fontweight='bold')
ax2.legend(fontsize=8, loc='lower right')
ax2.grid(True, alpha=0.3)

# 6.3. Середній retention
ax3 = axes[0, 2]
avg_retention_actual = predictions_df.groupby('lifetime_weeks')['retention_actual'].mean() * 100
avg_retention_pred = predictions_df.groupby('lifetime_weeks')['retention_predicted'].mean() * 100

weeks_actual = avg_retention_actual.dropna().index
ax3.plot(weeks_actual, avg_retention_actual.loc[weeks_actual], 
        'o-', linewidth=3, markersize=8, label='Actual', color='#2ecc71')
ax3.plot(avg_retention_pred.index, avg_retention_pred.values, 
        '--', linewidth=3, label='Predicted (sBG)', color='#e74c3c')

if len(weeks_actual) > 0:
    ax3.axvline(x=weeks_actual.max(), color='gray', linestyle=':', 
               label=f'Last Actual (Week {int(weeks_actual.max())})', linewidth=2)

ax3.set_xlabel('Lifetime (Weeks)', fontweight='bold')
ax3.set_ylabel('Retention (%)', fontweight='bold')
ax3.set_title('Average Retention: Actual vs Predicted', fontweight='bold')
ax3.legend(fontsize=9)
ax3.grid(True, alpha=0.3)
ax3.set_ylim(0, 105)

# 6.4. Середній LTV
ax4 = axes[1, 0]
avg_ltv_actual = predictions_df.groupby('lifetime_weeks')['ltv_actual'].mean()
avg_ltv_pred = predictions_df.groupby('lifetime_weeks')['ltv_predicted'].mean()

weeks_actual_ltv = avg_ltv_actual.dropna().index
ax4.plot(weeks_actual_ltv, avg_ltv_actual.loc[weeks_actual_ltv], 
        'o-', linewidth=3, markersize=8, label='Actual LTV', color='#2ecc71')
ax4.plot(avg_ltv_pred.index, avg_ltv_pred.values, 
        '--', linewidth=3, label='Predicted LTV', color='#e74c3c')

if len(weeks_actual_ltv) > 0:
    ax4.axvline(x=weeks_actual_ltv.max(), color='gray', linestyle=':', 
               label=f'Last Actual (Week {int(weeks_actual_ltv.max())})', linewidth=2)

ax4.set_xlabel('Lifetime (Weeks)', fontweight='bold')
ax4.set_ylabel('Average LTV ($)', fontweight='bold')
ax4.set_title('Average LTV: Actual vs Predicted', fontweight='bold')
ax4.legend(fontsize=9)
ax4.grid(True, alpha=0.3)

# 6.5. ARPU по тижнях
ax5 = axes[1, 1]
avg_arpu_actual = predictions_df.groupby('lifetime_weeks')['arpu_actual'].mean()
avg_arpu_pred = predictions_df.groupby('lifetime_weeks')['arpu_predicted'].mean()

weeks_arpu = avg_arpu_actual.dropna().index
if len(weeks_arpu) > 0:
    ax5.bar(weeks_arpu - 0.2, avg_arpu_actual.loc[weeks_arpu], 
           width=0.4, label='Actual ARPU', color='#3498db', alpha=0.7)
    ax5.bar(weeks_arpu + 0.2, avg_arpu_pred.loc[weeks_arpu], 
           width=0.4, label='Predicted ARPU', color='#e74c3c', alpha=0.7)

ax5.set_xlabel('Lifetime (Weeks)', fontweight='bold')
ax5.set_ylabel('ARPU per Week ($)', fontweight='bold')
ax5.set_title('ARPU by Week: Actual vs Predicted', fontweight='bold')
ax5.legend(fontsize=9)
ax5.grid(True, alpha=0.3, axis='y')

# 6.6. Actual vs Predicted LTV
ax6 = axes[1, 2]
if len(validation_data) > 0:
    scatter = ax6.scatter(validation_data['ltv_actual'], 
                         validation_data['ltv_predicted'], 
                         c=validation_data['lifetime_weeks'],
                         alpha=0.6, s=50, cmap='viridis')
    
    min_val = min(validation_data['ltv_actual'].min(), validation_data['ltv_predicted'].min())
    max_val = max(validation_data['ltv_actual'].max(), validation_data['ltv_predicted'].max())
    ax6.plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2, label='Perfect Prediction')
    
    ax6.set_xlabel('Actual LTV ($)', fontweight='bold')
    ax6.set_ylabel('Predicted LTV ($)', fontweight='bold')
    ax6.set_title(f'Actual vs Predicted (R²={r2:.3f})', fontweight='bold')
    
    cbar = plt.colorbar(scatter, ax=ax6)
    cbar.set_label('Week', fontweight='bold')
    ax6.legend(fontsize=9)
    ax6.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('ltv_analysis_charts.png', dpi=150, bbox_inches='tight')
print("Графіки збережено: ltv_analysis_charts.png")
plt.show()

# 7. ПРОГНОЗ НА 52 ТИЖНІ
print("\n\n7. ПРОГНОЗ LTV НА 52 ТИЖНІ")
print("-"*80)

week_52_data = predictions_df[predictions_df['lifetime_weeks'] == 52]
ltv_52 = week_52_data.groupby('cohort_week')['ltv_predicted'].mean()
retention_52 = week_52_data.groupby('cohort_week')['retention_predicted'].mean()
total_revenue_52 = (week_52_data['ltv_predicted'] * week_52_data['cohort_size']).sum()

print(f"\nПрогноз на 52 тижні:")
print(f"  Середній LTV: ${ltv_52.mean():.2f}")
print(f"  Медіанний LTV: ${ltv_52.median():.2f}")
print(f"  Діапазон LTV: ${ltv_52.min():.2f} - ${ltv_52.max():.2f}")
print(f"  Середній retention: {retention_52.mean()*100:.1f}%")
print(f"  Прогнозний сукупний дохід: ${total_revenue_52:,.2f}")

# Останній фактичний тиждень
last_week = predictions_df['ltv_actual'].last_valid_index()
if last_week is not None:
    last_week_num = predictions_df.loc[last_week, 'lifetime_weeks']
    avg_last_ltv = predictions_df[predictions_df['lifetime_weeks'] == last_week_num]['ltv_actual'].mean()
    growth = ((ltv_52.mean() - avg_last_ltv) / avg_last_ltv) * 100
    
    print(f"\nПорівняння:")
    print(f"  Останній фактичний LTV (тиждень {int(last_week_num)}): ${avg_last_ltv:.2f}")
    print(f"  Очікуваний приріст: ${ltv_52.mean() - avg_last_ltv:.2f} ({growth:+.1f}%)")
    
    # Аналіз чому такий приріст
    last_retention = predictions_df[predictions_df['lifetime_weeks'] == last_week_num]['retention_actual'].mean()
    print(f"\n  Пояснення:")
    print(f"    - Retention зараз: {last_retention*100:.1f}% → Прогноз на тиждень 52: {retention_52.mean()*100:.1f}%")
    print(f"    - Модель враховує природний churn користувачів")
    print(f"    - LTV зростає повільно через зменшення активних користувачів")

# 8. HTML ЗВІТ
print("\n\n8. ЗБЕРЕЖЕННЯ У HTML")
print("-"*80)

html_content = f"""<!DOCTYPE html>
<html lang="uk">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>LTV Prediction Analysis - Probabilistic Models — Hanna Todorova</title>
    <style>
        body {{
            font-family: Arial, sans-serif;
            margin: 20px;
            background-color: #f5f5f5;
        }}
        .container {{
            max-width: 1400px;
            margin: 0 auto;
            background-color: white;
            padding: 30px;
            box-shadow: 0 0 10px rgba(0,0,0,0.1);
        }}
        h1 {{
            color: #2c3e50;
            border-bottom: 3px solid #3498db;
            padding-bottom: 10px;
        }}
        h2 {{
            color: #34495e;
            margin-top: 30px;
            border-left: 4px solid #3498db;
            padding-left: 10px;
        }}
        h3 {{
            color: #34495e;
            margin-top: 20px;
        }}
        table {{
            border-collapse: collapse;
            width: 100%;
            margin: 20px 0;
            font-size: 14px;
        }}
        th, td {{
            border: 1px solid #ddd;
            padding: 12px;
            text-align: left;
        }}
        th {{
            background-color: #3498db;
            color: white;
            font-weight: bold;
        }}
        tr:nth-child(even) {{
            background-color: #f9f9f9;
        }}
        .metric {{
            display: inline-block;
            margin: 10px 20px 10px 0;
            padding: 15px;
            background-color: #ecf0f1;
            border-radius: 5px;
            min-width: 200px;
        }}
        .metric-label {{
            font-weight: bold;
            color: #7f8c8d;
            font-size: 14px;
        }}
        .metric-value {{
            font-size: 24px;
            color: #2c3e50;
            margin-top: 5px;
        }}
        .quality-excellent {{
            color: #27ae60;
            font-weight: bold;
        }}
        .quality-good {{
            color: #2ecc71;
            font-weight: bold;
        }}
        .quality-fair {{
            color: #f39c12;
            font-weight: bold;
        }}
        .quality-poor {{
            color: #e74c3c;
            font-weight: bold;
        }}
        img {{
            max-width: 100%;
            height: auto;
            margin: 20px 0;
            border: 1px solid #ddd;
        }}
        .info-box {{
            background-color: #e3f2fd;
            border-left: 5px solid #2196f3;
            padding: 15px;
            margin: 20px 0;
        }}
        .method-box {{
            background-color: #f3e5f5;
            border-left: 5px solid #9c27b0;
            padding: 15px;
            margin: 20px 0;
        }}
        .warning-box {{
            background-color: #fff3cd;
            border-left: 5px solid #ffc107;
            padding: 15px;
            margin: 20px 0;
        }}
        .formula {{
            background-color: #f8f9fa;
            padding: 10px;
            border-radius: 5px;
            font-family: 'Courier New', monospace;
            font-size: 14px;
            margin: 10px 0;
        }}
        ul {{
            line-height: 1.8;
        }}
        .highlight {{
            background-color: #fff9c4;
            padding: 2px 5px;
            border-radius: 3px;
        }}
    </style>
</head>
<body>
    <div class="container">
        <h1>📊 LTV Prediction Analysis - Probabilistic Models — Hanna Todorova</h1>
        <p><strong>Дата аналізу:</strong> {pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')}</p>
        
        <div class="warning-box">
            <h4>🎯 Ймовірнісний підхід до прогнозування LTV</h4>
            <p>Ця модель використовує <strong>sBG (Shifted Beta-Geometric)</strong> для retention та <strong>Gamma/BdW</strong> для монетизації:</p>
            
            <function_calls>
            <invoke name="artifacts">
            <parameter name="command">update</parameter>
            <parameter name="id">ltv_prediction_analysis</parameter>
            <parameter name="new_str">            <ul>
            <li><strong>sBG модель</strong> - моделює ймовірність churn кожного тижня</li>
            <li><strong>Gamma/BdW</strong> - враховує зменшення витрат активних користувачів</li>
            <li><strong>Реалістичний прогноз</strong> - LTV виходить на плато через churn</li>
            <li><strong>Математично обґрунтовано</strong> - базується на теорії ймовірностей</li>
            </ul>
        </div>
   
   <h2>1. Загальна інформація про дані</h2>
    <div class="metric">
        <div class="metric-label">Транзакцій</div>
        <div class="metric-value">{len(df):,}</div>
    </div>
    <div class="metric">
        <div class="metric-label">Користувачів</div>
        <div class="metric-value">{df['user_id'].nunique():,}</div>
    </div>
    <div class="metric">
        <div class="metric-label">Когорт</div>
        <div class="metric-value">{df['cohort_week'].nunique()}</div>
    </div>
    <div class="metric">
        <div class="metric-label">Макс lifetime</div>
        <div class="metric-value">{df['lifetime_weeks'].max()} тижнів</div>
    </div>
    
    <h2>2. Ймовірнісні моделі</h2>
    <div class="method-box">
        <h3>🔬 Модель 1: sBG (Shifted Beta-Geometric)</h3>
        <div class="formula">
        
        P(активний на тижні t) = Beta(α, β + t) / Beta(α, β)
        </div>
        <p><strong>Призначення:</strong> Моделювання retention та churn</p>
        <ul>
        <li><strong>α</strong> - параметр "липкості" продукту (вищий = краща retention)</li>
        <li><strong>β</strong> - параметр churn rate (нижчий = менший churn)</li>
        <li>Модель передбачає ймовірність того, що користувач залишиться активним</li>
        </ul>
        
         <h3>🔬 Модель 2: Gamma/BdW для монетизації</h3>
        <div class="formula">
        
        Revenue_per_period(t) = scale × (α_rev / (α_rev + t))
        </div>
        <p><strong>Призначення:</strong> Моделювання витрат активних користувачів</p>
        <ul>
        <li><strong>scale</strong> - базовий рівень витрат</li>
        <li><strong>α_rev</strong> - швидкість зменшення витрат з часом</li>
        <li>Враховує, що активні користувачі витрачають менше з часом</li>
        </ul>
        
         <h3>📐 Фінальний LTV</h3>
        <div class="formula">
        
        LTV(t) = Σ [Retention(i) × Revenue_per_active(i)] для i = 0..t
        </div>
    </div>
        
         <h3>Параметри моделей (приклади)</h3>
    <table>
        <tr>
            <th>Когорта</th>
            <th>α (retention)</th>
            <th>β (churn)</th>
            <th>α_rev</th>
            <th>scale ($)</th>
        </tr>
        
"""
for i, (cohort, params) in enumerate(list(model_params.items())[:10]):  
    html_content += f"""
<tr>
<td>{cohort.strftime('%Y-%m-%d')}</td>
<td>{params['alpha']:.3f}</td>
<td>{params['beta']:.3f}</td>
<td>{params['alpha_rev']:.3f}</td>
<td>${params['scale']:.2f}</td>
</tr>
"""
quality_class = 'quality-excellent' if r2 > 0.9 else 'quality-good' if r2 > 0.8 else 'quality-fair' if r2 > 0.7 else 'quality-poor'
quality_text = 'ВІДМІННА' if r2 > 0.9 else 'ХОРОША' if r2 > 0.8 else 'ЗАДОВІЛЬНА' if r2 > 0.7 else 'НИЗЬКА'
html_content += f"""
</table>
        
         <h2>3. Якість моделі</h2>
    <table>
        <tr>
            <th>Метрика</th>
            <th>Значення</th>
            <th>Опис</th>
        </tr>
        <tr>
            <td><strong>R² Score</strong></td>
            <td class="{quality_class}">{r2:.4f}</td>
            <td>Коефіцієнт детермінації</td>
        </tr>
        <tr>
            <td><strong>MAE</strong></td>
            <td>${mae:.2f}</td>
            <td>Середня абсолютна помилка</td>
        </tr>
        <tr>
            <td><strong>RMSE</strong></td>
            <td>${rmse:.2f}</td>
            <td>Корінь середньоквадратичної помилки</td>
        </tr>
        <tr>
            <td><strong>MAPE</strong></td>
            <td>{mape:.2f}%</td>
            <td>Середня абсолютна процентна помилка</td>
        </tr>
    </table>
    
    <div class="info-box">
        <h4 class="{quality_class}">✓ {quality_text} ЯКІСТЬ МОДЕЛІ (R² = {r2:.4f})</h4>
        <p>Модель пояснює {r2*100:.1f}% варіації в даних LTV.</p>
    </div>
    
    <h2>4. Прогноз LTV на 52 тижні</h2>
    <div class="metric">
        <div class="metric-label">Середній LTV (52 тижні)</div>
        <div class="metric-value">${ltv_52.mean():.2f}</div>
    </div>
    <div class="metric">
        <div class="metric-label">Медіанний LTV</div>
        <div class="metric-value">${ltv_52.median():.2f}</div>
    </div>
    <div class="metric">
        <div class="metric-label">Діапазон LTV</div>
        <div class="metric-value">${ltv_52.min():.2f} - ${ltv_52.max():.2f}</div>
    </div>
    <div class="metric">
        <div class="metric-label">Середній retention (52 тижні)</div>
        <div class="metric-value">{retention_52.mean()*100:.1f}%</div>
    </div>
    <div class="metric">
        <div class="metric-label">Прогнозний сукупний дохід</div>
        <div class="metric-value">${total_revenue_52:,.0f}</div>
    </div>
"""
if last_week is not None:
    html_content += f"""
<div class="info-box">
<h4>📊 Порівняння з фактичними даними</h4>
<p><strong>Останній фактичний LTV (тиждень {int(last_week_num)}):</strong> ${avg_last_ltv:.2f}</p>
<p><strong>Прогноз на 52 тижні:</strong> ${ltv_52.mean():.2f}</p>
<p><strong>Очікуваний приріст:</strong> ${ltv_52.mean() - avg_last_ltv:.2f} 
(<span class="highlight">{growth:+.1f}%</span>)</p>
<br>
<p><strong>Пояснення прогнозу:</strong></p>
<ul>
<li>Retention зараз: {last_retention*100:.1f}% → Прогноз: {retention_52.mean()*100:.1f}%</li>
<li>Модель враховує природний churn користувачів</li>
<li>LTV зростає повільно через зменшення активних користувачів</li>
<li>Ймовірнісна модель дає реалістичний прогноз без завищень</li>
</ul>
</div>

"""
cohort_predictions = week_52_data.groupby('cohort_week').agg({
    'ltv_predicted': 'mean',
    'retention_predicted': 'mean',
    'cohort_size': 'first'
}).reset_index()

cohort_predictions['total_revenue'] = (
    cohort_predictions['ltv_predicted'] * cohort_predictions['cohort_size']
)

cohort_predictions = cohort_predictions.sort_values('cohort_week')

html_content += """
<h3>Прогноз по когортах (52 тижні)</h3>
<table>
<tr>
<th>Когорта</th>
<th>Розмір</th>
<th>LTV (52 тижні)</th>
<th>Retention (52 тижні)</th>
<th>Сукупний revenue</th>
</tr>

"""
for _, row in cohort_predictions.head(15).iterrows():
    html_content += f"""
<tr>
<td>{row['cohort_week'].strftime('%Y-%m-%d')}</td>
<td>{int(row['cohort_size']):,}</td>
<td>${row['ltv_predicted']:.2f}</td>
<td>{row['retention_predicted']*100:.1f}%</td>
<td>${row['total_revenue']:,.0f}</td>
</tr>
"""
html_content += f"""
</table>
    
     <h2>5. Візуалізація</h2>
    <img src="ltv_analysis_charts.png" alt="LTV Analysis Charts">
    
    <h2>6. Висновки та рекомендації</h2>
    <div class="info-box">
        <h4>🎓 Ключові висновки:</h4>
        <ul>
            <li>Використано <strong>ймовірнісні моделі sBG + Gamma/BdW</strong> для реалістичного прогнозу</li>
            <li>Модель досягла якості <strong>R² = {r2:.4f}</strong></li>
            <li>Прогнозний LTV на рік: <strong>${ltv_52.mean():.2f}</strong></li>
            <li>Очікуваний retention через рік: <strong>{retention_52.mean()*100:.1f}%</strong></li>
            <li>Прогнозний сукупний дохід: <strong>${total_revenue_52:,.0f}</strong></li>
        </ul>
        
        <h4>💡 Рекомендації для покращення бізнесу:</h4>
        <ul>
            <li><strong>Покращити retention:</strong> Знизити churn через onboarding, engagement features</li>
            <li><strong>Збільшити ARPU:</strong> Upselling, premium features, персоналізація</li>
            <li><strong>Сегментація:</strong> Будувати окремі моделі для high-value сегментів</li>
            <li><strong>Моніторинг:</strong> Відстежувати retention та LTV по когортах щотижня</li>
            <li><strong>Експерименти:</strong> A/B тести для покращення retention на ранніх тижнях</li>
        </ul>
    </div>
    
    <div class="warning-box">
        <h4>⚠️ Обмеження моделі</h4>
        <ul>
            <li>Модель передбачає стабільність продукту (без великих змін)</li>
            <li>Не враховує сезонність та зовнішні фактори</li>
            <li>Точність знижується для дуже віддалених прогнозів (>1 року)</li>
            <li>Рекомендується оновлювати модель щомісяця з новими даними</li>
        </ul>
    </div>
</div>

</body>
</html>
"""
with open('ltv_prediction_report_4.html', 'w', encoding='utf-8') as f:
    f.write(html_content)
print("HTML звіт збережено: ltv_prediction_report_4.html")
print("\nДля перегляду відкрийте файл ltv_prediction_report_4.html у браузері")
print("\n" + "="*80)
```
