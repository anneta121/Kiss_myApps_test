```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# –ó–ê–í–ê–ù–¢–ê–ñ–ï–ù–ù–Ø –î–ê–ù–ò–•
df = pd.read_csv(r'C:\Users\user\Downloads\KMA TestTask Product Analyst - Task1.csv')


print("="*80)
print("Mobile App A/B Test Analysis ‚Äî Hanna Todorova")
print("="*80)

# 1. –ó–ê–ì–ê–õ–¨–ù–ò–ô –û–ì–õ–Ø–î –î–ê–ù–ò–•
print("\n1. –ó–ê–ì–ê–õ–¨–ù–ê –Ü–ù–§–û–†–ú–ê–¶–Ü–Ø –ü–†–û –î–ê–ù–Ü")
print("-"*80)
print(f"–ó–∞–≥–∞–ª—å–Ω–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á—ñ–≤: {len(df)}")
print(f"\n–†–æ–∑–ø–æ–¥—ñ–ª –ø–æ —Ç–µ—Å—Ç–æ–≤–∏–º –≥—Ä—É–ø–∞–º:")
print(df['test_group'].value_counts())
print(f"\n–†–æ–∑–ø–æ–¥—ñ–ª –ø–æ –∫—Ä–∞—ó–Ω–∞—Ö:")
print(df['country'].value_counts())

# –ü–ï–†–ï–í–Ü–†–ö–ê –¢–†–ò–í–ê–õ–û–°–¢–Ü –¢–ï–°–¢–£
print("\n\nüìÖ –¢–†–ò–í–ê–õ–Ü–°–¢–¨ –¢–ï–°–¢–£")
print("-"*80)
df['install_date'] = pd.to_datetime(df['install_date'])
min_date = df['install_date'].min()
max_date = df['install_date'].max()
test_duration = (max_date - min_date).days + 1  # +1 —â–æ–± –≤–∫–ª—é—á–∏—Ç–∏ –æ–±–∏–¥–≤–∞ –¥–Ω—ñ

print(f"–ü–æ—á–∞—Ç–æ–∫ —Ç–µ—Å—Ç—É: {min_date.strftime('%Y-%m-%d')}")
print(f"–ö—ñ–Ω–µ—Ü—å —Ç–µ—Å—Ç—É: {max_date.strftime('%Y-%m-%d')}")
print(f"–¢—Ä–∏–≤–∞–ª—ñ—Å—Ç—å: {test_duration} –¥–Ω—ñ–≤")

# –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü—ñ—ó –ø–æ —Ç—Ä–∏–≤–∞–ª–æ—Å—Ç—ñ
if test_duration < 7:
    print("‚ö†Ô∏è  –£–í–ê–ì–ê: –¢–µ—Å—Ç —Ç—Ä–∏–≤–∞—î –º–µ–Ω—à–µ 7 –¥–Ω—ñ–≤ - –ù–ï–î–û–°–¢–ê–¢–ù–¨–û!")
    print("   –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü—ñ—è: –º—ñ–Ω—ñ–º—É–º 1-2 —Ç–∏–∂–Ω—ñ –¥–ª—è –≤—Ä–∞—Ö—É–≤–∞–Ω–Ω—è —Ü–∏–∫–ª—ñ—á–Ω–æ—Å—Ç—ñ")
elif test_duration < 14:
    print("‚ö†Ô∏è  –ü–û–ü–ï–†–ï–î–ñ–ï–ù–ù–Ø: –¢–µ—Å—Ç —Ç—Ä–∏–≤–∞—î –º–µ–Ω—à–µ 2 —Ç–∏–∂–Ω—ñ–≤")
    print("   –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü—ñ—è: –±–∞–∂–∞–Ω–æ –ø—Ä–æ–¥–æ–≤–∂–∏—Ç–∏ –¥–æ 2-4 —Ç–∏–∂–Ω—ñ–≤")
elif test_duration < 28:
    print("‚úì  –ü–†–ò–ô–ù–Ø–¢–ù–û: –¢—Ä–∏–≤–∞–ª—ñ—Å—Ç—å —Ç–µ—Å—Ç—É –¥–æ—Å—Ç–∞—Ç–Ω—è (2-4 —Ç–∏–∂–Ω—ñ)")
else:
    print("‚úì  –í–Ü–î–ú–Ü–ù–ù–û: –¢–µ—Å—Ç —Ç—Ä–∏–≤–∞—î –¥–æ—Å—Ç–∞—Ç–Ω—å–æ –¥–æ–≤–≥–æ (>4 —Ç–∏–∂–Ω—ñ–≤)")

# –†–æ–∑–ø–æ–¥—ñ–ª –ø–æ –¥–Ω—è—Ö —Ç–∏–∂–Ω—è
print(f"\n–†–æ–∑–ø–æ–¥—ñ–ª —ñ–Ω—Å—Ç–∞–ª—ñ–≤ –ø–æ –¥–Ω—è—Ö —Ç–∏–∂–Ω—è:")
df['day_of_week'] = df['install_date'].dt.day_name()
day_distribution = df['day_of_week'].value_counts()
print(day_distribution)

# –ü–ï–†–ï–í–Ü–†–ö–ê –†–û–ó–ú–Ü–†–£ –í–ò–ë–Ü–†–ö–ò
print("\n\nüìä –ê–ù–ê–õ–Ü–ó –†–û–ó–ú–Ü–†–£ –í–ò–ë–Ü–†–ö–ò")
print("-"*80)

# 2. –û–°–ù–û–í–ù–Ü –ú–ï–¢–†–ò–ö–ò –ü–û –ì–†–£–ü–ê–•
print("\n\n2. –ü–û–†–Ü–í–ù–Ø–ù–ù–Ø –û–°–ù–û–í–ù–ò–• –ú–ï–¢–†–ò–ö")
print("-"*80)

metrics_summary = df.groupby('test_group').agg({
    'user_id': 'count',
    'trial': 'sum',
    'paid': 'sum',
    'revenue_1m': 'sum'
}).rename(columns={'user_id': 'total_users'})

metrics_summary['trial_rate'] = metrics_summary['trial'] / metrics_summary['total_users'] * 100
metrics_summary['paid_rate'] = metrics_summary['paid'] / metrics_summary['total_users'] * 100
metrics_summary['conversion_rate'] = metrics_summary['paid'] / metrics_summary['trial'] * 100
metrics_summary['arpu'] = metrics_summary['revenue_1m'] / metrics_summary['total_users']
metrics_summary['arppu'] = metrics_summary['revenue_1m'] / metrics_summary['paid']

print(metrics_summary.round(2))

# –†–û–ó–†–ê–•–£–ù–û–ö –ù–ï–û–ë–•–Ü–î–ù–û–ì–û –†–û–ó–ú–Ü–†–£ –í–ò–ë–Ü–†–ö–ò
from statsmodels.stats.power import zt_ind_solve_power

# ================================
# –†–û–ó–†–ê–•–£–ù–û–ö –†–û–ó–ú–Ü–†–£ –í–ò–ë–Ü–†–ö–ò
# ================================

groups = sorted(df['test_group'].unique())

control_name = [g for g in groups if "control" in g.lower()] or [groups[0]]
treatment_name = [g for g in groups if "treatment" in g.lower()] or [groups[1]]

group_a = df[df['test_group'] == control_name[0]]
group_b = df[df['test_group'] == treatment_name[0]]

baseline_conversion = group_a['paid'].mean()
treatment_conversion = group_b['paid'].mean()

effect_size_abs = treatment_conversion - baseline_conversion
effect_relative = (
    effect_size_abs / baseline_conversion * 100
    if baseline_conversion > 0 else 0
)

print("\nüìå –ë–ê–ó–û–í–Ü –ü–û–ö–ê–ó–ù–ò–ö–ò (Paid rate)\n")
print(f"Control ({control_name[0]}):  {baseline_conversion*100:.2f}%")
print(f"Treatment ({treatment_name[0]}): {treatment_conversion*100:.2f}%")
print(f"–°–ø–æ—Å—Ç–µ—Ä–µ–∂—É–≤–∞–Ω–∞ —Ä—ñ–∑–Ω–∏—Ü—è: {effect_size_abs*100:.2f} –ø.–ø. "
      f"({effect_relative:.1f}% –≤—ñ–¥–Ω–æ—Å–Ω–æ)")

# =====================================================
# –†–æ–∑—Ä–∞—Ö—É–Ω–æ–∫ –Ω–µ–æ–±—Ö—ñ–¥–Ω–æ–≥–æ n –¥–ª—è —Ä—ñ–∑–Ω–∏—Ö MDE (5%,10%,20%)
# =====================================================

print("\nüìê –ù–ï–û–ë–•–Ü–î–ù–ò–ô –†–û–ó–ú–Ü–† –í–ò–ë–Ü–†–ö–ò –ù–ê –ì–†–£–ü–£")
print("   (Œ± = 0.05, –ø–æ—Ç—É–∂–Ω—ñ—Å—Ç—å = 0.8)\n")

required_samples = {}

for relative_effect in [0.05, 0.10, 0.20]:

    expected_conversion = baseline_conversion * (1 + relative_effect)
    effect = expected_conversion - baseline_conversion

    # —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–æ–≤–∞–Ω–∏–π –µ—Ñ–µ–∫—Ç –¥–ª—è z-—Ç–µ—Å—Ç—É –ø—Ä–æ–ø–æ—Ä—Ü—ñ–π
    pooled_p = (baseline_conversion + expected_conversion) / 2
    pooled_se = np.sqrt(pooled_p * (1 - pooled_p))

    standardized_effect = effect / pooled_se if pooled_se > 0 else 0

    required_n = zt_ind_solve_power(
        effect_size=standardized_effect,
        alpha=0.05,
        power=0.8,
        ratio=1.0,
        alternative='two-sided'
    )

    required_samples[relative_effect] = int(required_n)

    print(f"  ‚ñ∂ {relative_effect*100:.0f}% –≤—ñ–¥–Ω–æ—Å–Ω–∞ –∑–º—ñ–Ω–∞: "
          f"{int(required_n):,} –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á—ñ–≤")

# —Ñ–∞–∫—Ç–∏—á–Ω–∞ –≤–∏–±—ñ—Ä–∫–∞
actual_n = min(len(group_a), len(group_b))

print(f"\nüìä –§–∞–∫—Ç–∏—á–Ω–∏–π —Ä–æ–∑–º—ñ—Ä –≤–∏–±—ñ—Ä–∫–∏: {actual_n:,} –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á—ñ–≤ –Ω–∞ –≥—Ä—É–ø—É\n")

# ======================
# –û—Ü—ñ–Ω–∫–∞ –¥–æ—Å—Ç–∞—Ç–Ω–æ—Å—Ç—ñ
# ======================

print("üí° –û–¶–Ü–ù–ö–ê –î–û–°–¢–ê–¢–ù–û–°–¢–Ü –î–ê–ù–ò–•:")

if actual_n >= required_samples[0.05]:
    print(" ‚úì –î–æ—Å—Ç–∞—Ç–Ω—å–æ –¥–ª—è –≤–∏—è–≤–ª–µ–Ω–Ω—è –º–∞–ª–∏—Ö –µ—Ñ–µ–∫—Ç—ñ–≤ (~5%)")
elif actual_n >= required_samples[0.10]:
    print(" ‚ö° –î–æ—Å—Ç–∞—Ç–Ω—å–æ –¥–ª—è —Å–µ—Ä–µ–¥–Ω—ñ—Ö –µ—Ñ–µ–∫—Ç—ñ–≤ (~10%)")
elif actual_n >= required_samples[0.20]:
    print(" ‚ö† –î–æ—Å—Ç–∞—Ç–Ω—å–æ –ª–∏—à–µ –¥–ª—è –≤–µ–ª–∏–∫–∏—Ö –µ—Ñ–µ–∫—Ç—ñ–≤ (~20%)")
else:
    print(" ‚ùå –í–∏–±—ñ—Ä–∫–∞ –∑–∞–º–∞–ª–∞ ‚Äî –≤–∏—Å–æ–∫–∏–π —Ä–∏–∑–∏–∫ –Ω–µ –≤–∏—è–≤–∏—Ç–∏ –µ—Ñ–µ–∫—Ç")
    
    # –ü–ï–†–ï–í–Ü–†–ö–ê –ë–ê–õ–ê–ù–°–£ –ì–†–£–ü
    print(f"\n‚öñÔ∏è  –ë–ê–õ–ê–ù–° –ú–Ü–ñ –ì–†–£–ü–ê–ú–ò:")
    ratio = len(group_a) / len(group_b) if len(group_b) > 0 else 0
    print(f"   –ì—Ä—É–ø–∞ {groups[0]}: {len(group_a):,} ({len(group_a)/len(df)*100:.1f}%)")
    print(f"   –ì—Ä—É–ø–∞ {groups[1]}: {len(group_b):,} ({len(group_b)/len(df)*100:.1f}%)")
    print(f"   –°–ø—ñ–≤–≤—ñ–¥–Ω–æ—à–µ–Ω–Ω—è: {ratio:.2f}")
    
    if 0.9 <= ratio <= 1.1:
        print("   ‚úì  –í–Ü–î–ú–Ü–ù–ù–û: –ì—Ä—É–ø–∏ –∑–±–∞–ª–∞–Ω—Å–æ–≤–∞–Ω—ñ (—Ä—ñ–∑–Ω–∏—Ü—è <10%)")
    elif 0.8 <= ratio <= 1.25:
        print("   ‚ö° –ü–†–ò–ô–ù–Ø–¢–ù–û: –ù–µ–≤–µ–ª–∏–∫–µ –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è –≤—ñ–¥ –±–∞–ª–∞–Ω—Å—É")
    else:
        print("   ‚ö†Ô∏è  –£–í–ê–ì–ê: –ó–Ω–∞—á–Ω–∏–π –¥–∏—Å–±–∞–ª–∞–Ω—Å –º—ñ–∂ –≥—Ä—É–ø–∞–º–∏!")
        print("   –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü—ñ—è: –ø–µ—Ä–µ–≤—ñ—Ä–∏—Ç–∏ —Ä–∞–Ω–¥–æ–º—ñ–∑–∞—Ü—ñ—é")

# 3. –°–¢–ê–¢–ò–°–¢–ò–ß–ù–Ü –¢–ï–°–¢–ò
print("\n\n3. –°–¢–ê–¢–ò–°–¢–ò–ß–ù–Ü –¢–ï–°–¢–ò (–ó–ê–ì–ê–õ–¨–ù–Ü)")
print("-"*80)

if len(groups) == 2:
    # Trial Rate Test
    trial_a = [group_a['trial'].sum(), len(group_a) - group_a['trial'].sum()]
    trial_b = [group_b['trial'].sum(), len(group_b) - group_b['trial'].sum()]
    chi2_trial, p_trial = stats.chi2_contingency([trial_a, trial_b])[:2]
    
    print(f"\nTrial Rate:")
    print(f"  –ì—Ä—É–ø–∞ {groups[0]}: {group_a['trial'].mean()*100:.2f}%")
    print(f"  –ì—Ä—É–ø–∞ {groups[1]}: {group_b['trial'].mean()*100:.2f}%")
    print(f"  Chi-square —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞: {chi2_trial:.4f}")
    print(f"  P-value: {p_trial:.4f}")
    print(f"  –†–µ–∑—É–ª—å—Ç–∞—Ç: {'–ó–ù–ê–ß–£–©–ò–ô' if p_trial < 0.05 else '–ù–ï –ó–ù–ê–ß–£–©–ò–ô'} (Œ±=0.05)")
    
    # Paid Rate Test
    paid_a = [group_a['paid'].sum(), len(group_a) - group_a['paid'].sum()]
    paid_b = [group_b['paid'].sum(), len(group_b) - group_b['paid'].sum()]
    chi2_paid, p_paid = stats.chi2_contingency([paid_a, paid_b])[:2]
    
    print(f"\nPaid Rate:")
    print(f"  –ì—Ä—É–ø–∞ {groups[0]}: {group_a['paid'].mean()*100:.2f}%")
    print(f"  –ì—Ä—É–ø–∞ {groups[1]}: {group_b['paid'].mean()*100:.2f}%")
    print(f"  Chi-square —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞: {chi2_paid:.4f}")
    print(f"  P-value: {p_paid:.4f}")
    print(f"  –†–µ–∑—É–ª—å—Ç–∞—Ç: {'–ó–ù–ê–ß–£–©–ò–ô' if p_paid < 0.05 else '–ù–ï –ó–ù–ê–ß–£–©–ò–ô'} (Œ±=0.05)")
    
    # Revenue Test (Mann-Whitney U test)
    stat_revenue, p_revenue = stats.mannwhitneyu(
        group_a['revenue_1m'].fillna(0), 
        group_b['revenue_1m'].fillna(0),
        alternative='two-sided'
    )
    
    print(f"\nRevenue (ARPU):")
    print(f"  –ì—Ä—É–ø–∞ {groups[0]}: ${group_a['revenue_1m'].mean():.2f}")
    print(f"  –ì—Ä—É–ø–∞ {groups[1]}: ${group_b['revenue_1m'].mean():.2f}")
    print(f"  Mann-Whitney U —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞: {stat_revenue:.4f}")
    print(f"  P-value: {p_revenue:.4f}")
    print(f"  –†–µ–∑—É–ª—å—Ç–∞—Ç: {'–ó–ù–ê–ß–£–©–ò–ô' if p_revenue < 0.05 else '–ù–ï –ó–ù–ê–ß–£–©–ò–ô'} (Œ±=0.05)")

# 4. –ê–ù–ê–õ–Ü–ó –ü–û –ö–†–ê–á–ù–ê–•
print("\n\n4. –ê–ù–ê–õ–Ü–ó –ü–û –ö–†–ê–á–ù–ê–•")
print("-"*80)

country_results = []

for country in df['country'].unique():
    df_country = df[df['country'] == country]
    
    if len(groups) == 2:
        country_a = df_country[df_country['test_group'] == groups[0]]
        country_b = df_country[df_country['test_group'] == groups[1]]
        
        # –ü–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ, —á–∏ –¥–æ—Å—Ç–∞—Ç–Ω—å–æ –¥–∞–Ω–∏—Ö
        if len(country_a) > 10 and len(country_b) > 10:
            # Trial rate
            trial_rate_a = country_a['trial'].mean() * 100
            trial_rate_b = country_b['trial'].mean() * 100
            trial_country_a = [country_a['trial'].sum(), len(country_a) - country_a['trial'].sum()]
            trial_country_b = [country_b['trial'].sum(), len(country_b) - country_b['trial'].sum()]
            
            try:
                _, p_trial_country = stats.chi2_contingency([trial_country_a, trial_country_b])[:2]
            except:
                p_trial_country = np.nan
            
            # Paid rate
            paid_rate_a = country_a['paid'].mean() * 100
            paid_rate_b = country_b['paid'].mean() * 100
            paid_country_a = [country_a['paid'].sum(), len(country_a) - country_a['paid'].sum()]
            paid_country_b = [country_b['paid'].sum(), len(country_b) - country_b['paid'].sum()]
            
            try:
                _, p_paid_country = stats.chi2_contingency([paid_country_a, paid_country_b])[:2]
            except:
                p_paid_country = np.nan
            
            # Revenue
            arpu_a = country_a['revenue_1m'].mean()
            arpu_b = country_b['revenue_1m'].mean()
            
            try:
                _, p_revenue_country = stats.mannwhitneyu(
                    country_a['revenue_1m'].fillna(0), 
                    country_b['revenue_1m'].fillna(0),
                    alternative='two-sided'
                )
            except:
                p_revenue_country = np.nan
            
            country_results.append({
                'country': country,
                'users': len(df_country),
                f'trial_rate_{groups[0]}': trial_rate_a,
                f'trial_rate_{groups[1]}': trial_rate_b,
                'trial_p_value': p_trial_country,
                'trial_significant': '–¢–ê–ö' if p_trial_country < 0.05 else '–ù–Ü',
                f'paid_rate_{groups[0]}': paid_rate_a,
                f'paid_rate_{groups[1]}': paid_rate_b,
                'paid_p_value': p_paid_country,
                'paid_significant': '–¢–ê–ö' if p_paid_country < 0.05 else '–ù–Ü',
                f'arpu_{groups[0]}': arpu_a,
                f'arpu_{groups[1]}': arpu_b,
                'revenue_p_value': p_revenue_country,
                'revenue_significant': '–¢–ê–ö' if p_revenue_country < 0.05 else '–ù–Ü'
            })

country_df = pd.DataFrame(country_results).sort_values('users', ascending=False)
print("\n–î–µ—Ç–∞–ª—å–Ω–∞ —Ç–∞–±–ª–∏—Ü—è –ø–æ –∫—Ä–∞—ó–Ω–∞—Ö:")
print(country_df.to_string(index=False))

# 5. –í–Ü–ó–£–ê–õ–Ü–ó–ê–¶–Ü–Ø
print("\n\n5. –í–Ü–ó–£–ê–õ–Ü–ó–ê–¶–Ü–Ø")
print("-"*80)

fig = plt.figure(figsize=(16, 12))
gs = fig.add_gridspec(2, 2, hspace=0.25, wspace=0.25, top=0.95, bottom=0.05, left=0.08, right=0.98)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

fig.suptitle('A/B Test Analysis Dashboard', fontsize=16, fontweight='bold')

# 5.1. Trial Rate –ø–æ –≥—Ä—É–ø–∞—Ö
trial_data = df.groupby('test_group')['trial'].mean() * 100
ax1.bar(trial_data.index, trial_data.values, color=['#3498db', '#e74c3c'])
ax1.set_title('Trial Rate by Test Group', fontweight='bold')
ax1.set_ylabel('Trial Rate (%)', fontweight='bold')
ax1.set_xlabel('Test Group', fontweight='bold')
for p in ax1.patches:
    value = p.get_height()
    ax1.text(
        p.get_x() + p.get_width() / 2,
        value * 0.5,
        f'{value:.2f}%',
        ha='center',
        va='center',
        color='white',
        fontweight='bold'
    )

    # 5.2. Paid Rate –ø–æ –≥—Ä—É–ø–∞—Ö
paid_data = df.groupby('test_group')['paid'].mean() * 100
ax2.bar(paid_data.index, paid_data.values, color=['#3498db', '#e74c3c'])
ax2.set_title('Paid Rate by Test Group', fontweight='bold')
ax2.set_ylabel('Paid Rate (%)', fontweight='bold')
ax2.set_xlabel('Test Group', fontweight='bold')
for p in ax2.patches:
    value = p.get_height()
    ax2.text(
        p.get_x() + p.get_width() / 2,
        value * 0.5,
        f'{value:.2f}%',
        ha='center',
        va='center',
        color='white',
        fontweight='bold'
    )
    
# 5.3. ARPU –ø–æ –≥—Ä—É–ø–∞—Ö
arpu_data = df.groupby('test_group')['revenue_1m'].mean()
ax3.bar(arpu_data.index, arpu_data.values, color=['#3498db', '#e74c3c'])
ax3.set_title('ARPU by Test Group', fontweight='bold')
ax3.set_ylabel('ARPU ($)', fontweight='bold')
ax3.set_xlabel('Test Group', fontweight='bold')
for p in ax3.patches:
    value = p.get_height()
    ax3.text(
        p.get_x() + p.get_width() / 2,
        value * 0.5,
        f'{value:.2f}%',
        ha='center',
        va='center',
        color='white',
        fontweight='bold'
    )

# 5.4. Paid Rate –ø–æ –∫—Ä–∞—ó–Ω–∞—Ö
country_paid = df.groupby(['country', 'test_group'])['paid'].mean().unstack() * 100
country_paid.plot(kind='bar', ax=ax4, color=['#3498db', '#e74c3c'])
ax4.set_title('Paid Rate by Country and Test Group', fontweight='bold')
ax4.set_ylabel('Paid Rate (%)', fontweight='bold')
ax4.set_xlabel('Country', fontweight='bold')
legend = ax4.legend(title='Test Group')
legend.get_title().set_fontweight('bold')
ax4.tick_params(axis='x', rotation=45)
for p in ax4.patches:
    value = p.get_height()
    ax4.text(
        p.get_x() + p.get_width() / 2,
        value * 0.5,
        f'{value:.2f}%',
        ha='center',
        va='center',
        color='white',
        fontweight='bold',
        fontsize=8 
    )

plt.show()

# 6. –í–ò–°–ù–û–í–ö–ò
print("\n\n6. –í–ò–°–ù–û–í–ö–ò –¢–ê –†–ï–ö–û–ú–ï–ù–î–ê–¶–Ü–á")
print("-"*80)

if len(groups) == 2:
    winner = groups[1] if metrics_summary.loc[groups[1], 'paid_rate'] > metrics_summary.loc[groups[0], 'paid_rate'] else groups[0]
    
    print(f"\n–ü–µ—Ä–µ–º–æ–∂–µ—Ü—å –∑–∞ Paid Rate: –ì—Ä—É–ø–∞ {winner}")
    print(f"\n–°—Ç–∞—Ç–∏—Å—Ç–∏—á–Ω–∞ –∑–Ω–∞—á—É—â—ñ—Å—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤:")
    print(f"  - Trial Rate: {'–ó–ù–ê–ß–£–©–ò–ô' if p_trial < 0.05 else '–ù–ï –ó–ù–ê–ß–£–©–ò–ô'}")
    print(f"  - Paid Rate: {'–ó–ù–ê–ß–£–©–ò–ô' if p_paid < 0.05 else '–ù–ï –ó–ù–ê–ß–£–©–ò–ô'}")
    print(f"  - Revenue: {'–ó–ù–ê–ß–£–©–ò–ô' if p_revenue < 0.05 else '–ù–ï –ó–ù–ê–ß–£–©–ò–ô'}")
    
    significant_countries = country_df[country_df['paid_significant'] == '–¢–ê–ö']['country'].tolist()
    if significant_countries:
        print(f"\n–ö—Ä–∞—ó–Ω–∏ –∑—ñ –∑–Ω–∞—á—É—â–∏–º–∏ –≤—ñ–¥–º—ñ–Ω–Ω–æ—Å—Ç—è–º–∏: {', '.join(significant_countries)}")
    else:
        print("\n–ù–µ–º–∞—î –∫—Ä–∞—ó–Ω –∑—ñ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–Ω–æ –∑–Ω–∞—á—É—â–∏–º–∏ –≤—ñ–¥–º—ñ–Ω–Ω–æ—Å—Ç—è–º–∏")

print("\n" + "="*80)

# 7. –ó–ë–ï–†–ï–ñ–ï–ù–ù–Ø –†–ï–ó–£–õ–¨–¢–ê–¢–Ü–í –£ HTML
print("\n\n7. –ó–ë–ï–†–ï–ñ–ï–ù–ù–Ø –†–ï–ó–£–õ–¨–¢–ê–¢–Ü–í")
print("-"*80)

fig.savefig('ab_test_charts.png', dpi=150, bbox_inches='tight')
print("–ì—Ä–∞—Ñ—ñ–∫–∏ –∑–±–µ—Ä–µ–∂–µ–Ω–æ: ab_test_charts.png")

# –°—Ç–≤–æ—Ä—é—î–º–æ HTML –∑–≤—ñ—Ç
html_content = """<!DOCTYPE html>
<html lang="uk">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>A/B Test Analysis Report (Mobile App) ‚Äî Hanna Todorova</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            margin: 20px;
            background-color: #f5f5f5;
        }
        .container {
            max-width: 1200px;
            margin: 0 auto;
            background-color: white;
            padding: 30px;
            box-shadow: 0 0 10px rgba(0,0,0,0.1);
        }
        h1 {
            color: #2c3e50;
            border-bottom: 3px solid #3498db;
            padding-bottom: 10px;
        }
        h2 {
            color: #34495e;
            margin-top: 30px;
            border-left: 4px solid #3498db;
            padding-left: 10px;
        }
        h3 {
            color: #34495e;
            margin-top: 20px;
        }
        table {
            border-collapse: collapse;
            width: 100%;
            margin: 20px 0;
            font-size: 14px;
        }
        th, td {
            border: 1px solid #ddd;
            padding: 12px;
            text-align: left;
        }
        th {
            background-color: #3498db;
            color: white;
            font-weight: bold;
        }
        tr:nth-child(even) {
            background-color: #f9f9f9;
        }
        .metric {
            display: inline-block;
            margin: 10px 20px 10px 0;
            padding: 15px;
            background-color: #ecf0f1;
            border-radius: 5px;
            min-width: 200px;
        }
        .metric-label {
            font-weight: bold;
            color: #7f8c8d;
            font-size: 14px;
        }
        .metric-value {
            font-size: 24px;
            color: #2c3e50;
            margin-top: 5px;
        }
        .significant {
            color: #27ae60;
            font-weight: bold;
        }
        .not-significant {
            color: #e74c3c;
            font-weight: bold;
        }
        .winner {
            background-color: #2ecc71;
            color: white;
            padding: 10px 20px;
            border-radius: 5px;
            display: inline-block;
            font-weight: bold;
            font-size: 18px;
        }
        img {
            max-width: 100%;
            height: auto;
            margin: 20px 0;
            border: 1px solid #ddd;
        }
        .summary-box {
            background-color: #fff3cd;
            border-left: 5px solid #ffc107;
            padding: 15px;
            margin: 20px 0;
        }
        .validity-box {
            background-color: #e8f5e9;
            border-left: 5px solid #4caf50;
            padding: 15px;
            margin: 20px 0;
        }
        .warning-box {
            background-color: #ffebee;
            border-left: 5px solid #f44336;
            padding: 15px;
            margin: 20px 0;
        }
        .info-box {
            background-color: #e3f2fd;
            border-left: 5px solid #2196f3;
            padding: 15px;
            margin: 20px 0;
        }
        .status-icon {
            font-size: 20px;
            margin-right: 10px;
        }
    </style>
</head>
<body>
    <div class="container">
"""

html_content += f"""
        <h1>üìä A/B Test Analysis Report (Mobile App) ‚Äî Hanna Todorova</h1>
        <p><strong>–î–∞—Ç–∞ –∞–Ω–∞–ª—ñ–∑—É:</strong> {pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')}</p>
        
        <h2>üîç –í–∞–ª—ñ–¥–Ω—ñ—Å—Ç—å —Ç–∞ –Ω–∞–¥—ñ–π–Ω—ñ—Å—Ç—å –¥–∞–Ω–∏—Ö</h2>
        
        <h3>üìÖ –¢—Ä–∏–≤–∞–ª—ñ—Å—Ç—å —Ç–µ—Å—Ç—É</h3>
        <div class="metric">
            <div class="metric-label">–ü–æ—á–∞—Ç–æ–∫ —Ç–µ—Å—Ç—É</div>
            <div class="metric-value">{min_date.strftime('%Y-%m-%d')}</div>
        </div>
        <div class="metric">
            <div class="metric-label">–ö—ñ–Ω–µ—Ü—å —Ç–µ—Å—Ç—É</div>
            <div class="metric-value">{max_date.strftime('%Y-%m-%d')}</div>
        </div>
        <div class="metric">
            <div class="metric-label">–¢—Ä–∏–≤–∞–ª—ñ—Å—Ç—å</div>
            <div class="metric-value">{test_duration} –¥–Ω—ñ–≤</div>
        </div>
"""

# –û—Ü—ñ–Ω–∫–∞ —Ç—Ä–∏–≤–∞–ª–æ—Å—Ç—ñ
if test_duration < 7:
    html_content += """
        <div class="warning-box">
            <span class="status-icon">‚ö†Ô∏è</span>
            <strong>–£–í–ê–ì–ê: –¢–µ—Å—Ç —Ç—Ä–∏–≤–∞—î –º–µ–Ω—à–µ 7 –¥–Ω—ñ–≤ - –ù–ï–î–û–°–¢–ê–¢–ù–¨–û!</strong><br>
            –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü—ñ—è: –º—ñ–Ω—ñ–º—É–º 1-2 —Ç–∏–∂–Ω—ñ –¥–ª—è –≤—Ä–∞—Ö—É–≤–∞–Ω–Ω—è —Ü–∏–∫–ª—ñ—á–Ω–æ—Å—Ç—ñ
        </div>
"""
elif test_duration < 14:
    html_content += """
        <div class="info-box">
            <span class="status-icon">‚ö†Ô∏è</span>
            <strong>–ü–û–ü–ï–†–ï–î–ñ–ï–ù–ù–Ø: –¢–µ—Å—Ç —Ç—Ä–∏–≤–∞—î –º–µ–Ω—à–µ 2 —Ç–∏–∂–Ω—ñ–≤</strong><br>
            –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü—ñ—è: –±–∞–∂–∞–Ω–æ –ø—Ä–æ–¥–æ–≤–∂–∏—Ç–∏ –¥–æ 2-4 —Ç–∏–∂–Ω—ñ–≤
        </div>
"""
elif test_duration < 28:
    html_content += """
        <div class="validity-box">
            <span class="status-icon">‚úì</span>
            <strong>–ü–†–ò–ô–ù–Ø–¢–ù–û: –¢—Ä–∏–≤–∞–ª—ñ—Å—Ç—å —Ç–µ—Å—Ç—É –¥–æ—Å—Ç–∞—Ç–Ω—è (2-4 —Ç–∏–∂–Ω—ñ)</strong>
        </div>
"""
else:
    html_content += """
        <div class="validity-box">
            <span class="status-icon">‚úì</span>
            <strong>–í–Ü–î–ú–Ü–ù–ù–û: –¢–µ—Å—Ç —Ç—Ä–∏–≤–∞—î –¥–æ—Å—Ç–∞—Ç–Ω—å–æ –¥–æ–≤–≥–æ (>4 —Ç–∏–∂–Ω—ñ–≤)</strong>
        </div>
"""

if len(groups) == 2:
    actual_n = min(len(group_a), len(group_b))
    html_content += f"""
        <h3>üìä –†–æ–∑–º—ñ—Ä –≤–∏–±—ñ—Ä–∫–∏</h3>
        <div class="metric">
            <div class="metric-label">–ì—Ä—É–ø–∞ {groups[0]}</div>
            <div class="metric-value">{len(group_a):,}</div>
        </div>
        <div class="metric">
            <div class="metric-label">–ì—Ä—É–ø–∞ {groups[1]}</div>
            <div class="metric-value">{len(group_b):,}</div>
        </div>
        <div class="metric">
            <div class="metric-label">–°–ø—ñ–≤–≤—ñ–¥–Ω–æ—à–µ–Ω–Ω—è –≥—Ä—É–ø</div>
            <div class="metric-value">{ratio:.2f}</div>
        </div>
"""
    
    # –û—Ü—ñ–Ω–∫–∞ —Ä–æ–∑–º—ñ—Ä—É –≤–∏–±—ñ—Ä–∫–∏
    if actual_n >= required_samples[0.05]:
    #print(" ‚úì –î–æ—Å—Ç–∞—Ç–Ω—å–æ –¥–ª—è –≤–∏—è–≤–ª–µ–Ω–Ω—è –º–∞–ª–∏—Ö –µ—Ñ–µ–∫—Ç—ñ–≤ (~5%)"):
        html_content += """
        <div class="warning-box">
            <span class="status-icon">‚úì</span>
            <strong>–î–æ—Å—Ç–∞—Ç–Ω—å–æ –¥–ª—è –≤–∏—è–≤–ª–µ–Ω–Ω—è –º–∞–ª–∏—Ö –µ—Ñ–µ–∫—Ç—ñ–≤ (~5%)</strong><br>
            –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü—ñ—è: –∑—ñ–±—Ä–∞—Ç–∏ –∑–Ω–∞—á–Ω–æ –±—ñ–ª—å—à–µ –¥–∞–Ω–∏—Ö
        </div>
"""      
    elif actual_n >= required_samples[0.10]:
        html_content += """
        <div class="warning-box">
            <span class="status-icon">‚ö°</span>
            <strong>–î–æ—Å—Ç–∞—Ç–Ω—å–æ –¥–ª—è —Å–µ—Ä–µ–¥–Ω—ñ—Ö –µ—Ñ–µ–∫—Ç—ñ–≤ (~10%)</strong><br>
            –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü—ñ—è: –ø—Ä–æ–¥–æ–≤–∂–∏—Ç–∏ —Ç–µ—Å—Ç –¥–ª—è –Ω–∞–¥—ñ–π–Ω–∏—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤
        </div>
"""
    elif actual_n >= required_samples[0.20]:
        html_content += """
        <div class="info-box">
            <span class="status-icon">‚ö†</span>
            <strong>–î–æ—Å—Ç–∞—Ç–Ω—å–æ –ª–∏—à–µ –¥–ª—è –≤–µ–ª–∏–∫–∏—Ö –µ—Ñ–µ–∫—Ç—ñ–≤ (~20%)</strong><br>
            –î–ª—è –º–∞–ª–∏—Ö –µ—Ñ–µ–∫—Ç—ñ–≤ (<5%) –ø–æ—Ç—Ä—ñ–±–Ω–æ –±—ñ–ª—å—à–µ –¥–∞–Ω–∏—Ö
        </div>
       
"""
    else:
        html_content += """
        <div class="validity-box">
            <span class="status-icon">‚ùå</span>
            <strong>–í–∏–±—ñ—Ä–∫–∞ –∑–∞–º–∞–ª–∞ ‚Äî –≤–∏—Å–æ–∫–∏–π —Ä–∏–∑–∏–∫ –Ω–µ –≤–∏—è–≤–∏—Ç–∏ –µ—Ñ–µ–∫—Ç</strong>
        </div>
"""
    
    # –û—Ü—ñ–Ω–∫–∞ –±–∞–ª–∞–Ω—Å—É
    if 0.9 <= ratio <= 1.1:
        html_content += """
        <div class="validity-box">
            <span class="status-icon">‚úì</span>
            <strong>–í–Ü–î–ú–Ü–ù–ù–û: –ì—Ä—É–ø–∏ –∑–±–∞–ª–∞–Ω—Å–æ–≤–∞–Ω—ñ (—Ä—ñ–∑–Ω–∏—Ü—è <10%)</strong>
        </div>
"""
    elif 0.8 <= ratio <= 1.25:
        html_content += """
        <div class="info-box">
            <span class="status-icon">‚ö°</span>
            <strong>–ü–†–ò–ô–ù–Ø–¢–ù–û: –ù–µ–≤–µ–ª–∏–∫–µ –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è –≤—ñ–¥ –±–∞–ª–∞–Ω—Å—É</strong>
        </div>
"""
    else:
        html_content += """
        <div class="warning-box">
            <span class="status-icon">‚ö†Ô∏è</span>
            <strong>–£–í–ê–ì–ê: –ó–Ω–∞—á–Ω–∏–π –¥–∏—Å–±–∞–ª–∞–Ω—Å –º—ñ–∂ –≥—Ä—É–ø–∞–º–∏!</strong><br>
            –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü—ñ—è: –ø–µ—Ä–µ–≤—ñ—Ä–∏—Ç–∏ —Ä–∞–Ω–¥–æ–º—ñ–∑–∞—Ü—ñ—é
        </div>
"""

html_content += f"""
        <h2>1. –ó–∞–≥–∞–ª—å–Ω–∞ —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—è</h2>
        <div class="metric">
            <div class="metric-label">–ó–∞–≥–∞–ª—å–Ω–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á—ñ–≤</div>
            <div class="metric-value">{len(df):,}</div>
        </div>
                       
        <h2>2. –û—Å–Ω–æ–≤–Ω—ñ –º–µ—Ç—Ä–∏–∫–∏ –ø–æ –≥—Ä—É–ø–∞—Ö</h2>
        {metrics_summary.to_html(classes='table')}
"""

if len(groups) == 2:
    html_content += f"""
        <h2>3. –°—Ç–∞—Ç–∏—Å—Ç–∏—á–Ω—ñ —Ç–µ—Å—Ç–∏</h2>
        
        <div class="summary-box">
            <h3>–ü–µ—Ä–µ–º–æ–∂–µ—Ü—å: <span class="winner">–ì—Ä—É–ø–∞ {winner}</span></h3>
        </div>
        
        <h3>Trial Rate</h3>
        <table>
            <tr>
                <th>–ì—Ä—É–ø–∞</th>
                <th>Trial Rate</th>
                <th>Chi-square</th>
                <th>P-value</th>
                <th>–ó–Ω–∞—á—É—â—ñ—Å—Ç—å</th>
            </tr>
            <tr>
                <td>{groups[0]}</td>
                <td>{group_a['trial'].mean()*100:.2f}%</td>
                <td rowspan="2">{chi2_trial:.4f}</td>
                <td rowspan="2">{p_trial:.4f}</td>
                <td rowspan="2" class="{'significant' if p_trial < 0.05 else 'not-significant'}">
                    {'‚úì –ó–ù–ê–ß–£–©–ò–ô' if p_trial < 0.05 else '‚úó –ù–ï –ó–ù–ê–ß–£–©–ò–ô'}
                </td>
            </tr>
            <tr>
                <td>{groups[1]}</td>
                <td>{group_b['trial'].mean()*100:.2f}%</td>
            </tr>
        </table>
        
        <h3>Paid Rate</h3>
        <table>
            <tr>
                <th>–ì—Ä—É–ø–∞</th>
                <th>Paid Rate</th>
                <th>Chi-square</th>
                <th>P-value</th>
                <th>–ó–Ω–∞—á—É—â—ñ—Å—Ç—å</th>
            </tr>
            <tr>
                <td>{groups[0]}</td>
                <td>{group_a['paid'].mean()*100:.2f}%</td>
                <td rowspan="2">{chi2_paid:.4f}</td>
                <td rowspan="2">{p_paid:.4f}</td>
                <td rowspan="2" class="{'significant' if p_paid < 0.05 else 'not-significant'}">
                    {'‚úì –ó–ù–ê–ß–£–©–ò–ô' if p_paid < 0.05 else '‚úó –ù–ï –ó–ù–ê–ß–£–©–ò–ô'}
                </td>
            </tr>
            <tr>
                <td>{groups[1]}</td>
                <td>{group_b['paid'].mean()*100:.2f}%</td>
            </tr>
        </table>
        
        <h3>Revenue (ARPU)</h3>
        <table>
            <tr>
                <th>–ì—Ä—É–ø–∞</th>
                <th>ARPU</th>
                <th>Mann-Whitney U</th>
                <th>P-value</th>
                <th>–ó–Ω–∞—á—É—â—ñ—Å—Ç—å</th>
            </tr>
            <tr>
                <td>{groups[0]}</td>
                <td>${group_a['revenue_1m'].mean():.2f}</td>
                <td rowspan="2">{stat_revenue:.4f}</td>
                <td rowspan="2">{p_revenue:.4f}</td>
                <td rowspan="2" class="{'significant' if p_revenue < 0.05 else 'not-significant'}">
                    {'‚úì –ó–ù–ê–ß–£–©–ò–ô' if p_revenue < 0.05 else '‚úó –ù–ï –ó–ù–ê–ß–£–©–ò–ô'}
                </td>
            </tr>
            <tr>
                <td>{groups[1]}</td>
                <td>${group_b['revenue_1m'].mean():.2f}</td>
            </tr>
        </table>
"""
    
    # –†–æ–∑–±–∏–≤–∞—î–º–æ —Ç–∞–±–ª–∏—Ü—é –∫—Ä–∞—ó–Ω
    country_df_part1 = country_df[['country', 'users', 
                                     f'trial_rate_{groups[0]}', f'trial_rate_{groups[1]}', 
                                     'trial_p_value', 'trial_significant']].copy()
    country_df_part2 = country_df[['country', 'paid_significant',
                                     f'arpu_{groups[0]}', f'arpu_{groups[1]}',
                                     'revenue_p_value', 'revenue_significant']].copy()
    
    html_content += f"""
        <h2>4. –ê–Ω–∞–ª—ñ–∑ –ø–æ –∫—Ä–∞—ó–Ω–∞—Ö</h2>
        <h3>–ß–∞—Å—Ç–∏–Ω–∞ 1: Trial Rate –ø–æ –∫—Ä–∞—ó–Ω–∞—Ö</h3>
        {country_df_part1.to_html(classes='table', index=False)}
        
        <h3>–ß–∞—Å—Ç–∏–Ω–∞ 2: Revenue –ø–æ –∫—Ä–∞—ó–Ω–∞—Ö</h3>
        {country_df_part2.to_html(classes='table', index=False)}
"""

html_content += """
        <h2>5. –í—ñ–∑—É–∞–ª—ñ–∑–∞—Ü—ñ—è</h2>
        <img src="ab_test_charts.png" alt="A/B Test Charts">
        
        <h2>6. –í–∏—Å–Ω–æ–≤–∫–∏</h2>
        <div class="summary-box">
"""

if len(groups) == 2:
    html_content += f"""
            <p><strong>–ü–µ—Ä–µ–º–æ–∂–µ—Ü—å –∑–∞ Paid Rate:</strong> –ì—Ä—É–ø–∞ {winner}</p>
            <p><strong>–°—Ç–∞—Ç–∏—Å—Ç–∏—á–Ω–∞ –∑–Ω–∞—á—É—â—ñ—Å—Ç—å:</strong></p>
            <ul>
                <li>Trial Rate: <span class="{'significant' if p_trial < 0.05 else 'not-significant'}">{'–ó–ù–ê–ß–£–©–ò–ô' if p_trial < 0.05 else '–ù–ï –ó–ù–ê–ß–£–©–ò–ô'}</span></li>
                <li>Paid Rate: <span class="{'significant' if p_paid < 0.05 else 'not-significant'}">{'–ó–ù–ê–ß–£–©–ò–ô' if p_paid < 0.05 else '–ù–ï –ó–ù–ê–ß–£–©–ò–ô'}</span></li>
                <li>Revenue: <span class="{'significant' if p_revenue < 0.05 else 'not-significant'}">{'–ó–ù–ê–ß–£–©–ò–ô' if p_revenue < 0.05 else '–ù–ï –ó–ù–ê–ß–£–©–ò–ô'}</span></li>
            </ul>
"""
    
    significant_countries = country_df[country_df['paid_significant'] == '–¢–ê–ö']['country'].tolist()
    if significant_countries:
        html_content += f"<p><strong>–ö—Ä–∞—ó–Ω–∏ –∑—ñ –∑–Ω–∞—á—É—â–∏–º–∏ –≤—ñ–¥–º—ñ–Ω–Ω–æ—Å—Ç—è–º–∏:</strong> {', '.join(significant_countries)}</p>"
    else:
        html_content += "<p><strong>–ö—Ä–∞—ó–Ω–∏ –∑—ñ –∑–Ω–∞—á—É—â–∏–º–∏ –≤—ñ–¥–º—ñ–Ω–Ω–æ—Å—Ç—è–º–∏:</strong> –ù–µ–º–∞—î</p>"

html_content += """
        </div>
<style>
.conclusion-box {
    background-color: #e8f6ec;      /* —Å–≤—ñ—Ç–ª–æ-–∑–µ–ª–µ–Ω–∏–π —Ñ–æ–Ω */
    border-left: 5px solid #2f8f46; /* —Ç–æ–≤—Å—Ç–∞ –∑–µ–ª–µ–Ω–∞ —Å–º—É–≥–∞ –∑–ª—ñ–≤–∞ */
    padding: 16px 18px;
    margin-top: 16px;
    font-size: 15px;
}
.conclusion-box h3 {
    color: #2f8f46;
    margin-top: 0;
}
</style>
        <h2>7.–ü—Ä–æ–ø–æ–∑–∏—Ü—ñ—ó</h2>
        <div class="conclusion-box">


          <p>
            –ù–∞ –æ—Å–Ω–æ–≤—ñ –æ—Ç—Ä–∏–º–∞–Ω–∏—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤ A/B-—Ç–µ—Å—Ç—É —Ä–µ–∫–æ–º–µ–Ω–¥–æ–≤–∞–Ω–æ –ø—Ä–æ–≤–µ—Å—Ç–∏
            –ø–æ–≤—Ç–æ—Ä–Ω–∏–π –µ–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç, –æ—Å–∫—ñ–ª—å–∫–∏ –ø–æ—Ç–æ—á–Ω–∏–π —Ç–µ—Å—Ç –Ω–µ –¥–æ—Å—è–≥–Ω—É–≤
            —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–Ω–æ—ó –∑–Ω–∞—á—É—â–æ—Å—Ç—ñ —Ç–∞ –º—ñ–≥ –º–∞—Ç–∏ –Ω–µ–¥–æ—Å—Ç–∞—Ç–Ω—é —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–Ω—É –ø–æ—Ç—É–∂–Ω—ñ—Å—Ç—å.
          </p>

      <ul>
        <li>
          –ù–µ –∑—É–ø–∏–Ω—è—Ç–∏ –ø–æ–≤—Ç–æ—Ä–Ω–∏–π —Ç–µ—Å—Ç —Ä–∞–Ω—ñ—à–µ –Ω—ñ–∂ —á–µ—Ä–µ–∑ <strong>14 –¥–Ω—ñ–≤</strong>,
          —â–æ–± –ø–æ–∫—Ä–∏—Ç–∏ —Ç–∏–∂–Ω–µ–≤—É —Å–µ–∑–æ–Ω–Ω—ñ—Å—Ç—å —Ç–∞ –ø–æ–≤–µ–¥—ñ–Ω–∫–æ–≤—ñ –ø–∞—Ç–µ—Ä–Ω–∏ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á—ñ–≤.
        </li>

        <li>
          –ó–∞–±–µ–∑–ø–µ—á–∏—Ç–∏ <strong>–ø—Ä–æ–ø–æ—Ä—Ü—ñ–π–Ω–∏–π —Ä–æ–∑–ø–æ–¥—ñ–ª –≤–∏–±—ñ—Ä–∫–∏ –ø–æ –∫—Ä–∞—ó–Ω–∞—Ö</strong>,
          —â–æ–± —É–Ω–∏–∫–Ω—É—Ç–∏ –ø–µ—Ä–µ–∫–æ—Å—É —Å—Ç—Ä—É–∫—Ç—É—Ä–∏ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á—ñ–≤ –º—ñ–∂ –≥—Ä—É–ø–∞–º–∏.
        </li>

        <li>
          –¶—ñ–ª—å–æ–≤–∏–π —Ä–æ–∑–º—ñ—Ä –≤–∏–±—ñ—Ä–∫–∏ ‚Äî
          <strong>28 000 –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á—ñ–≤ –Ω–∞ –∫–æ–∂–Ω—É –≥—Ä—É–ø—É</strong>,
          —â–æ –∑–∞–±–µ–∑–ø–µ—á–∏—Ç—å –¥–æ—Å—Ç–∞—Ç–Ω—é —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–Ω—É –ø–æ—Ç—É–∂–Ω—ñ—Å—Ç—å –¥–ª—è –≤–∏—è–≤–ª–µ–Ω–Ω—è
          –ø–æ–º—ñ—Ä–Ω–∏—Ö –µ—Ñ–µ–∫—Ç—ñ–≤.
        </li>
      </ul>

      <p>
        –ü—ñ—Å–ª—è –∑–∞–≤–µ—Ä—à–µ–Ω–Ω—è –ø–µ—Ä—ñ–æ–¥—É –∑–±–æ—Ä—É –¥–∞–Ω–∏—Ö —Ä–µ–∫–æ–º–µ–Ω–¥—É—î—Ç—å—Å—è –ø–æ–≤—Ç–æ—Ä–Ω–æ
        –ø–µ—Ä–µ–≤—ñ—Ä–∏—Ç–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–Ω—É –∑–Ω–∞—á—É—â—ñ—Å—Ç—å, –≤–µ–ª–∏—á–∏–Ω—É –µ—Ñ–µ–∫—Ç—É —Ç–∞ —Å—Ç–∞–±—ñ–ª—å–Ω—ñ—Å—Ç—å
        –º–µ—Ç—Ä–∏–∫ —É –¥–∏–Ω–∞–º—ñ—Ü—ñ.
      </p>
        </div>
    </div>
</body>
</html>
"""

with open('ab_test_report.html', 'w', encoding='utf-8') as f:
    f.write(html_content)

print("HTML –∑–≤—ñ—Ç –∑–±–µ—Ä–µ–∂–µ–Ω–æ: ab_test_report.html")
print("\n–î–ª—è –ø–µ—Ä–µ–≥–ª—è–¥—É –≤—ñ–¥–∫—Ä–∏–π—Ç–µ —Ñ–∞–π–ª ab_test_report.html —É –±—Ä–∞—É–∑–µ—Ä—ñ")
```
